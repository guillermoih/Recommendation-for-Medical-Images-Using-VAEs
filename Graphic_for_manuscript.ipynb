# Plotting functions

```python
def acc_plot(k, acc_1, acc_2, acc_3, acc_4, acc_5):
    fig, ax = plt.subplots(1, figsize=(7,7), dpi=200)

    ax.plot(k, acc_1, 'o-', label='gamma=1')
    ax.plot(k, acc_2, 'x-', label='gamma=0.1')
    ax.plot(k, acc_3, '^-', label='gamma=0.01')
    ax.plot(k, acc_4, 's-', label='gamma=0.001')
    ax.plot(k, acc_5, 'D-', label='gamma=0.0001')
    ax.set_xticks(k)
    ax.set_xlabel('k')
    ax.set_ylabel('Precision@k')
    ax.set_ylim([35, 55])
    ax.legend(loc='upper right')

    plt.show()
```

```python
def wass_plot(k, wass_1, wass_2, wass_3, wass_4, wass_5):
    fig, ax = plt.subplots(1, figsize=(7,7), dpi=200)

    ax.plot(k, wass_1, 'o-', label='gamma=1')
    ax.plot(k, wass_2, 'x-', label='gamma=0.1')
    ax.plot(k, wass_3, '^-', label='gamma=0.01')
    ax.plot(k, wass_4, 's-', label='gamma=0.001')
    ax.plot(k, wass_5, 'D-', label='gamma=0.0001')
    ax.set_xticks(k)
    ax.set_xlabel('k')
    ax.set_ylabel('Sliced Wasserstein distance')
    ax.set_ylim([0.40, 0.51])
    ax.legend(loc='lower right')

    plt.show()
```

```python
k = [3, 4, 5, 10]
```

# AE and Classifier

```python
def acc_plot_1(k, acc_classifier, acc_ae):
    fig, ax = plt.subplots(1, figsize=(7,7), dpi=200)

    ax.plot(k, acc_classifier, 'o-', label='Classifier')
    ax.plot(k, acc_ae, 'x-', label='AE')
    ax.set_xticks(k)
    ax.set_xlabel('k')
    ax.set_ylabel('Precision@k')
    ax.set_ylim([35, 55])
    ax.legend(loc='upper right')

    plt.show()
```

```python
def wass_plot_1(k, wass_classifier, wass_ae):
    fig, ax = plt.subplots(1, figsize=(7,7), dpi=200)

    ax.plot(k, wass_classifier, 'o-', label='Classifier')
    ax.plot(k, wass_ae, 'x-', label='AE')
    ax.set_xticks(k)
    ax.set_xlabel('k')
    ax.set_ylabel('Sliced Wasserstein distance')
    ax.set_ylim([0.40, 0.51])
    ax.legend(loc='lower right')

    plt.show()
```

```python
classifier_acc = [50.47, 49.93, 48.69, 46.76]
ae_acc = [38.22, 37.17, 36.76, 35.13]

classifier_wass = [0.4702, 0.4787, 0.4848, 0.5074]
ae_wass = [0.4223, 0.4319, 0.4417, 0.4744]
```

```python
acc_plot_1(k, classifier_acc, ae_acc)
```

```python
wass_plot_1(k, classifier_wass, ae_wass)
```
